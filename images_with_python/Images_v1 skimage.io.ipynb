#### # replacing black pixels with white

```python

image = io.imread(os.path.join(input_dir, 'sample_image_dog_1.jpg'))

r,g,b = image[:,:,0], image[:,:,1], image[:,:,2]
mask = (r == 0) & (g == 0) & (b == 0)
image[:,:,:][mask] = [255,255,255]
plt.imshow(image)
```
